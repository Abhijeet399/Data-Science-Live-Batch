```python
x_cor = np.array([0,1,2])
plt.bar(x_cor, [10, 20, 15], width = 0.5, label = 'Blue data', color = 'aqua')
plt.bar(x_cor+0.5, [20, 10, 13], width=0.5, color = 'red', label = 'Red data', tick_label = ['x', 'y', 'z'])
plt.ylim(0, 30)
plt.xlim(-1, 3.5)
plt.xlabel('Time')
plt.ylabel('Sales')
plt.legend()
plt.title('Sundar Bar plot')
```
